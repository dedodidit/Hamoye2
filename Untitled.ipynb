I will like to see the top 20 Items with the highest mean values

```python
plt.figure(figsize=(10, 5))

df.groupby(['Item'])['Value'].mean().sort_values(ascending=False).head(20).plot(kind='barh')
```

I will liketo see the countries that produce the most Items

```python
df.groupby(['Country'])['Value'].sum().sort_values(ascending=False).head(20).plot(kind='barh')
```

Nigeria has the highest production value, followed by Egypt and South Africa

I would now check for what Item Nigeria produces the most

```python
nig = df[df['Country'] == 'Nigeria']
```

```python
nig.groupby(['Item'])['Value'].sum().sort_values(ascending=False).head(20).plot(kind='barh')
```

Cassava and products is the highest Item Nigeria produces, followed by Yams and Beverages

Overall, I'd like to see the highest produces from Africa
